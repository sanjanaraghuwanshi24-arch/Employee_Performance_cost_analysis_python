```python
#Task 2 : FILLING MISSING VALUE IN COST COLUMN USING RUNNING AVERAGE #USE PF LOOP AS REQUIRED 
The cost column in the dataframe “Project” has some missing values. Your task is to 
compute these missing values. Replace the missing values by running average. You 
should use the “For” loop for this task.
```

```python

pd.options.display.float_format = '{:.0f}'.format

# Fill missing Cost using running average

running_sum = 0
count = 0

for i in range(len(project_df)):
    
    # If value is NOT missing
    if pd.notna(project_df.loc[i, "Cost"]):
        running_sum += project_df.loc[i, "Cost"]
        count += 1
    
    # If value is missing
    else:
        project_df.loc[i, "Cost"] = running_sum / count

# Show updated dataframe
project_df
```

```python
#Task 6 
Demote the designation level by 1, whose projects have status “fail”. Delete the 
employees record whose designation level is above 4. 
```

```python
final_df = pd.merge(project_df, employee_df, on="ID")
final_df = pd.merge(final_df, seniority_df, on="ID")
```

```python
final_df.rename(columns={"Designation Level": "Designation"}, inplace=True)
```

```python
# Demote for failed projects
final_df.loc[final_df["Status"] == "Failed", "Designation"] += 1

# Remove designation > 4
final_df = final_df[final_df["Designation"] <= 4]

# Show result
final_df
```

```python
final_df["Bonus"] = np.where(final_df["Status"] == "Finished",
                            final_df["Cost"] * 0.05,
                            0)

final_df
```

```python
Task 9 
Add the cost of all projects for each Employee and save it in new dataframe 
“TotalProjCost” with three columns ID, First Name, and Total cost # calculating total project cost for each  employee
```

```python
# Group by ID and First Name, then sum cost
TotalProjCost = final_df.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

# Rename column
TotalProjCost.rename(columns={"Cost": "Total Cost"}, inplace=True)

# Show result
TotalProjCost
```
